```python
import os
os.environ["TRANSFORMERS_NO_TF"] = "1"
os.environ["TRANSFORMERS_NO_FLAX"] = "1"

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM


# Model ID for Mistral
model_id = "tiiuae/falcon-7b-instruct"


# Load model/tokenizer with authentication
tokenizer = AutoTokenizer.from_pretrained(model_id, use_auth_token=HF_TOKEN)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto",
    use_auth_token=HF_TOKEN
)

# System prompt
SYSTEM_PROMPT = (
    "You are a helpful and friendly medical assistant. "
    "You provide general health-related information in a safe and responsible way. "
    "Do not give any diagnosis or treatment. Always recommend consulting a healthcare professional."
)

def format_prompt(user_question):
    return f"<s>[INST] {SYSTEM_PROMPT}\n\n{user_question} [/INST]"

def ask_chatbot(question):
    prompt = format_prompt(question)
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=300,
        do_sample=True,
        temperature=0.6,
        top_p=0.95
    )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    final_response = response.split("[/INST]")[-1].strip()
    disclaimer = "\n\n‚ö†Ô∏è *Disclaimer: This information is not a substitute for professional medical advice. Always consult a doctor for medical concerns.*"
    return final_response + disclaimer

def main():
    print("ü©∫ Health Chatbot (Mistral-7B with Token Access)")
    print("Type 'exit' to quit.\n")
    while True:
        user_input = input("Enter your health question: ")
        if user_input.lower() == "exit":
            print("Goodbye! Stay safe ü©∫")
            break
        print("\nChatbot:", ask_chatbot(user_input), "\n")

if __name__ == "__main__":
    main()
```
